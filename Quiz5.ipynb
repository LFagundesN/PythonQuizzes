Q1: What happens when you join without a common column or index?
<br>
It tries to join by index, but if there’s no shared index and columns overlap, it gives an error.
<br>
Also:
<br>
employee_df contains employee information (e.g. Name, Department, etc.)
<br>
books_df contains book information (e.g. Title, Genre, etc.)
<br>
There’s no natural relationship between the two.
<br>
<br>
Q2: Why is this join not useful?
<br>
Because there's no key to connect the data, and pandas can’t figure out how to line up the rows. You need a proper index or column to join on.



```python
# Add a shared key column called Book_Index (1, 2, 3, ...)
books_df['Book_Index'] = range(1, len(books_df) + 1)
employee_df['Book_Index'] = range(1, len(employee_df) + 1)

# Set 'Book_Index' as the index in both DataFrames
books_df.set_index('Book_Index', inplace=True)
employee_df.set_index('Book_Index', inplace=True)
```

Both DataFrames had a column named 'Status', but they meant different things:
<br>
<br> - In employee_df: Status means employment status (e.g. Active, Terminated)
<br> - In books_df: Status means book availability (e.g. Available, Checked Out)
<br>
<br> When we joined them, pandas renamed the duplicate columns to avoid confusion:
<br>
<br> - Status_emp = from employee_df
<br> - Status_book = from books_df
<br>
This avoids a conflict and keeps both versions.

```python
# Perform left join (keep all employees)
#GPT HELPED ME WITH THAT
left_join = employee_df.join(books_df, how='left', lsuffix='_emp', rsuffix='_book')
```

```python
# Perform inner join (only where both sides have matching index)
#GPT HELPED ME WITH THAT
inner_join = employee_df.join(books_df, how='inner', lsuffix='_emp', rsuffix='_book')
```
